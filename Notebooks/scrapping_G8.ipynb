```python
# import libraries
from bs4 import BeautifulSoup
import urllib
import urllib.request
import csv
from tqdm import tqdm
import re
```

```python
#  l'url de notre requête
urlpage = 'https://pubmed.ncbi.nlm.nih.gov/?term=%28artificial+intelligence+OR+machine+learning+OR+deep+learning+OR+neural+network+OR+big+data%29AND%28image+recognition+diagnostic+OR+detecting+OR+drug+design+%29+AND%28cardiovascular+disease+OR+cancer+OR+asthma+OR+alzheimer+OR+diabetes+OR+AIDS+OR+schizophrenia+OR+Neurodegenerative+Diseases%29'
```

```python
recherches = []
```

```python
#Chaque numéro i(recherches) contient 10 articles
#On récupère donc le code de chaque numero i

#ici i va de 200 à 300 donc on scrappe le code de 1000 articles

i=1
for i in tqdm(range(200,300)):
    urlpagebis = urlpage + '&page=' + str(i)
    page = urllib.request.urlopen(urlpagebis)
    stock = BeautifulSoup(page, 'html.parser')
    recherches.append(stock)
    i+=1
```

```python
#Chaque numéro i (recherche) contient 10 articles
#on parcours donc chaque numero i pour extraire l'id tous les articles
#qu'il contient

b = []
for rech in recherches :
    for a in rech.find_all("a",href = True) :
        for c in re.findall('/[0-9]+/',a['href']):
            b.append(c)
```

```python
#on crée url de chaque article
url = 'https://pubmed.ncbi.nlm.nih.gov'
bis = [url + str(element) + '#affiliation-1' for element in b]
```

```python
#ici on parcourt chaque article (variable page et soup)
#en même temps on extrait la partie qui concerne l'auteur (variable em_box)

# On procède ainsi pour tous les éléments qui nous intéressent (date, mots-clés, ...)

auteur = []
for i in tqdm(range(0,len(bis))) :
    page = urllib.request.urlopen(bis[i])

    soup = BeautifulSoup(page, 'html5lib')
    em_box = soup.find("div", {"class": "article-page"})

    em_box = em_box.find("div", {"class": "affiliations"})

    auteur.append(em_box)
```
